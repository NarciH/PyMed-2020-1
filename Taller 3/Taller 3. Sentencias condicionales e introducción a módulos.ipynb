Dadas las coordenadas de un punto(x,y) decir si este punto está dentro, en o fuera de una circunferencia de radio dada (r).


```python
# r_c = radio_cordenada
'''
3 cosas que hacer.
Leer x,y, r
r_c = x^2+y^2
if r_c<=r:
    Imprimir Estoy dentro del circulo
else:
    Imprimir Estoy por fuera del circulo
'''

```

```python
x = float(input("Ingrese x: "))
y = float(input("Ingrese y: "))
r = float(input("Ingrese r: "))

r_c = (x**2+y**2)**(1/2)
if r_c == r:
    print('Imprimir Estoy dentro del circulo')
else:
    print('Imprimir Estoy por fuera del circulo')
```

# Solucionar numéricamente una ecuación diferencial con scipy

$$ \frac{dy}{dx} = \frac{1}{x}+\left(\frac{y}{x}\right)^2$$


```python
import numpy as np
from scipy.integrate import odeint
import plotly.express as px

def df(x,y): return 1/x+(y/x)**2
x=np.linspace(1,10,500)
y=odeint(df,0.1,x)
px.line(x = x, y = y)
```

### Ciclos dobles

```python
for i in [1,2,3]:
    list_new = []
    for j in ['a','b']:
        list_new.append(j+'_{}'.format(i))
    print(list_new)
```

```python
# !pip install itertools
from itertools import product

A = [1,2,3]
B = ['a','b']

for a,b in product(A,B):
    print(a,b)
```

```python
for a,b in zip(A,B):
    print(a,b)
```

# Ejemplo de interpretación de ejercicios

Realice un ciclo iterativo
imprima el nombre de cada uno de los archivos contenidos en la variable lista_archivos. En cada iteración el programa debe:
- Imprimir el nombre del archivo
- Realizar un ciclo iterativo que inicia en 3, va hasta 30 con un paso de 3. 


```python
lista_archivos = ['file_1.xlsx','file_2.xlsx','file_3.xlsx']
for archivo in lista_archivos:
    print(archivo)
    for i in range(3,30,3):
        print('\t',i)
```

```python
for archivo,i in product(lista_archivos,range(3,30,3)):
    print(archivo, i)
```
